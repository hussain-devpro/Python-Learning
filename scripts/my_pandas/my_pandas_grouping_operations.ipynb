There are 3 Types of grouping operation on Groups.

1. Aggregation
2. Transformation
3. Filtering

```python
import numpy as np
import pandas as pd
import os
```

```python
PICKLE_PATH = os.path.join('pandas_dataset', 'artwork_pickle_data.csv')
df = pd.read_pickle(PICKLE_PATH)
```

```python
min(df['acquisitionYear']), df['acquisitionYear'].min()  # both results are same
```

```python
small_df = df.iloc[49980:50019, :].copy()
np.count_nonzero(small_df)
```

1. Aggregation Example

```python
grouped = small_df.groupby('artist')
type(grouped)
```

```python
for name, group_df in grouped:
    print(name)
    print(group_df)
```

```python
for name, group_df in grouped:
    min_year = group_df['acquisitionYear'].min()
    print(f'{name}: {min_year}')
```

2. Transformation: Changing missing value with most common values

```python
group_medium = small_df.groupby('artist')['medium']
small_df.loc[:, 'medium'] = group_medium.transform(len)  # here we transform the mediym wiht its length function for str
# we can use user-defined functions in transformation as well
small_df
```
