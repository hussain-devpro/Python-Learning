```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import pandas as pd
```

```python
app = dash.Dash()
```

```python
# df = pd.read_csv('https://gist.githubusercontent.com/chriddyp/c78bf172206ce24f77d6363a2d754b59/raw/'
#                  'c353e8ef842413cae56ae3920b8fd78468aa4cb2/usa-agricultural-exports-2011.csv')
df = pd.read_csv('dash_dataset/usa-agricultural-exports-2011.csv')
df.head()
```

```python
def generated_html_table_from_dataframe(selected_df, max_rows=10):
    html_table = html.Table([
        html.Thead(
            html.Tr([html.Th(col) for col in selected_df.columns])
        ),
        html.Tbody([
            html.Tr([
                html.Td(selected_df.loc[i, col]) for col in selected_df.columns
            ]) for i in range(min(len(selected_df), max_rows))
        ])
    ], title = 'Table with a Title as tooltip')  # Title will aprear as tooltip for entire table
    return html_table
```

```python
app.layout = html.Div([
    html.H1('USA Agriculture export data!!!'),
    generated_html_table_from_dataframe(df)
])
```

```python
if __name__ == '__main__':
    app.run_server()
```
