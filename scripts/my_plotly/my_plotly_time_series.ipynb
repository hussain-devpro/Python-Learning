Time-series helps us spot trends in any industry, (specifically financial industries)

```python
from datetime import datetime
import plotly.graph_objs as go
import plotly.offline as offline
import pandas as pd
from pandas_datareader import DataReader
offline.init_notebook_mode(connected=True)
```

```python
google = DataReader('GOOGL', 'yahoo',
                      datetime(2017, 1, 1),
                      datetime(2018, 6, 1)).reset_index()
google.head()
```

```python
trace_high = go.Scatter(x = google.loc[:, 'Date'],
                        y = google.loc[:, 'High'],
                        name = 'Google High',
                        line = dict(color = '#6699FF')
                       )
trace_low = go.Scatter(x = google.loc[:, 'Date'],
                        y = google.loc[:, 'Low'],
                        name = 'Google Low',
                        line = dict(color = '#FF6633')
                       )
data = [trace_high, trace_low]

layout = go.Layout(title='Google Stock Price Data',
                   xaxis=dict(autorange=True,
                              rangeselector=dict(buttons=list([
                                  dict(count=1,
                                       label="1m",
                                       step="month",
                                       stepmode="backward"),
                                  dict(count=6,
                                       label="6m",
                                       step="month",
                                       stepmode="backward"),
                                  dict(count=1,
                                       label="YTD",
                                       step="year",
                                       stepmode="todate"),
                                  dict(count=1,
                                       label="1y",
                                       step="year",
                                       stepmode="backward"),
                                  dict(step="all")
                              ])),
                              rangeslider=dict(visible=True), # For more granular view we specify rangeslider (currently not working) on date
                              type="date"
                              )
                   )
google_fig = dict(data=data, layout=layout)
offline.iplot(data)
```
